#**One Hot/Dummy Encoding**

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ml basics/agora.csv")
df.head()
```

```python
pd.get_dummies(df['Area'])
```

```python
dummy_variables = pd.get_dummies(df['Area'],drop_first=True)  #drop_first deletes the first col(ctg)
                                                              #otherwise by deafult there would be (ctg Dhaka	Rangpur) 3 columns
```

```python
dummy_variables.head()
```

```python
df=pd.concat([df,dummy_variables], axis=1)    #joining encoded cols with dataframe
```

```python
df=df.drop('Area',axis=1)   #deleting unencoded col
```

```python
df.head()
```

#Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ml basics/agora.csv")
df.head()
```

```python
df.Area= df.Area.fillna('Dhaka')
```

```python
df.Area.unique()
```

```python
cities=['Dhaka', 'Ctg', 'Rangpur']
```

```python
ordinal = OrdinalEncoder(categories=[cities])
```

```python
encoded_val= ordinal.fit_transform(df[['Area']])
```

```python
new_area= pd.DataFrame(encoded_val)
```

```python
df.head() #area column still unencoded 
```

```python
df.Area=new_area  #replacing with encoded values
```

```python
df.head() #encoded
```
